## **Ejercicio Extra: COVID-19**

```python
# Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Filtrar los datos de México y asegurar que la columna 'date' sea tipo fecha (to_datetime)
df_mex = df[df['location'] == 'Mexico'].copy()
df_mex['date'] = pd.to_datetime(df_mex['date'])
df_mex.head()
```

```python
# Agregar un nueva columna (nuevas_muertes_diarias)
df_mex['nuevas_Muertes_Diarias'] = df_mex['total_deaths'].diff()
df_mex.head()
```

```python
# Buscar el día con más muertes diarias
pico = df_mex.loc[df_mex['nuevas_Muertes_Diarias'].idxmax()]
fecha_pico = pico['date']
muertes_pico = pico['nuevas_Muertes_Diarias']

# Total de muertes acumuladas al final del periodo
total_muertes = df_mex['total_deaths'].iloc[-1]

# Total de casos en MEX
casos_totales = df_mex['total_cases'].iloc[-1]
fecha_final = df_mex['date'].iloc[-1].date()

print(f"Casos confirmados totales en México hasta el {fecha_final}: {int(casos_totales)}")
print(f"Fecha con más muertes diarias: {fecha_pico.date()} - Muertes: {int(muertes_pico)}")
print(f"Total de muertes acumuladas al final del periodo: {int(total_muertes)}")
```

```python
df_mex.tail()
```

# **Evolución de Nuevas Muertes Diarias por COVID-19 en México**

```python
# Graficar la evolución de nuevas muertes diarias
plt.figure(figsize=(12, 6))
plt.plot(df_mex['date'], df_mex['nuevas_Muertes_Diarias'], label='Nuevas Muertes Diarias', color='blue')
plt.scatter(fecha_pico, muertes_pico, color='green', label='Pico de Muertes', zorder=5)
plt.title("Evolución de Nuevas Muertes Diarias por COVID-19 en México")
plt.xlabel("Fecha")
plt.ylabel("Muertes Diarias")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# **Evolución de Casos Acumulados de COVID-19 en México**

```python
# Grafica la evolución de casos acumulados
plt.figure(figsize=(12, 6))
plt.plot(df_mex['date'], df_mex['total_cases'], color='red', linewidth=2)
plt.title("Evolución de Casos Acumulados de COVID-19 en México")
plt.xlabel("Fecha")
plt.ylabel("Total de Casos Confirmados")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Porcentaje de Población Vacunada vs No Vacunada en México**

```python
# Eliminar filas con NaN en las columnas necesarias
df_mex = df_mex.dropna(subset=['people_fully_vaccinated', 'population'])

# Obtener el último dato disponible sin NaN
ultimo_dato = df_mex.iloc[-1]
vacunados = ultimo_dato['people_fully_vaccinated']
poblacion = ultimo_dato['population']
no_vacunados = poblacion - vacunados

# Verificar que no hay NaN
if pd.isna(vacunados) or pd.isna(no_vacunados):
    print("No hay datos completos para graficar.")
else:
    # grafica de pastel
    etiquetas = ['Vacunados', 'No vacunados']
    valores = [vacunados, no_vacunados]
    colores = ['green', 'red']

    plt.figure(figsize=(7, 7))
    plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', colors=colores, startangle=100)
    plt.title("Porcentaje de Población Vacunada vs No Vacunada en México")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```
