## **Ejercicio 1:**

```python
a = int(input("Ingrese el primer número (a): "))
b = int(input("Ingrese el segundo número (b): "))

# División entera
division_entera = a // b

# Resto de la división
resto = a % b

# Potencia a elevado a b
potencia = a ** b

print(f"División entera de {a} entre {b}: {division_entera}")
print(f"Resto de la división de {a} entre {b}: {resto}")
print(f"{a} elevado a la {b} es: {potencia}")
```

## **Ejercicio 3:**

```python
dic_divisas = {
  'Euro': '€',
  'Dollar': '$',
  'Yen': '¥'
}

nombre_divisa = input("Ingrese el nombre de una divisa (Euro, Dollar o Yen): ")

# Verificar si la divisa está en el diccionario y mostrar el resultado
if nombre_divisa in dic_divisas:
    print(f"El símbolo de {nombre_divisa} es: {dic_divisas[nombre_divisa]}")
else:
    print("La divisa no está en el diccionario.")
```

## **Ejercicio 4:**

```python
import matplotlib.pyplot as plt

n = int(input("Ingrese un número: "))

# sucesión de Fibonacci
fibonacci = [0, 1]
for i in range(2, n):
    siguiente = fibonacci[-1] + fibonacci[-2]
    fibonacci.append(siguiente)

print("Sucesión de Fibonacci:")
print(fibonacci)

# Gráfica de líneas
plt.plot(fibonacci, marker='o', linestyle='-', color='Red')
plt.title("Grafica: Crecimiento de la Sucesión de Fibonacci")
plt.xlabel("Posición en la Sucesión")
plt.ylabel("Valor de Fibonacci")
plt.grid(True)
plt.show()
```

## **Ejercicio 5:**

```python
import pandas as pd

anio_inicio = int(input("Ingrese el año de inicio del rango: "))
anio_fin = int(input("Ingrese el año de fin del rango: "))

anios = list(range(anio_inicio, anio_fin + 1))
ventas = []

# Solicitar las ventas para cada año
for anio in anios:
    venta = float(input(f"Ingrese las ventas del año {anio}: "))
    ventas.append(venta)

# Crear una Serie con los datos originales
serie_ventas = pd.Series(ventas, index=anios)

# Aplicar el descuento del 10%
serie_descuento = serie_ventas * 0.9

print("-" * 50)
print("Ventas originales:")
for anio, venta in serie_ventas.items():
    print(f"Año {anio}: {venta}")

print("-" * 50)
print("Ventas después de aplicar un descuento del 10%:")
for anio, venta in serie_descuento.items():
    print(f"Año {anio}: {venta}")
```
